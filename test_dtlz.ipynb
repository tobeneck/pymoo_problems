# Test if evaluated PS fits PF

```python
from moo.dtlz import DTLZ1, DTLZ2, DTLZ3, DTLZ5, DTLZ6
from helper import plot_front

n_paretos = 500
n_obj = 3

problem = DTLZ1(n_obj=n_obj)
nds = problem.evaluate(problem._calc_pareto_set(n_paretos))
plot_front(problem, nds, title="DTLZ1").show()

problem = DTLZ2(n_obj=n_obj)
nds = problem.evaluate(problem._calc_pareto_set(n_paretos))
plot_front(problem, nds, title="DTLZ2").show()

problem = DTLZ3(n_obj=n_obj)
nds = problem.evaluate(problem._calc_pareto_set(n_paretos))
plot_front(problem, nds, title="DTLZ3").show()

problem = DTLZ5(n_obj=n_obj)
nds = problem.evaluate(problem._calc_pareto_set(n_paretos))
plot_front(problem, nds, title="DTLZ5").show()

problem = DTLZ6(n_obj=n_obj)
nds = problem.evaluate(problem._calc_pareto_set(n_paretos))
plot_front(problem, nds, title="DTLZ6").show()
```
