# Test the pareto sets

The results of PF and PS are visually compared to the reference PS from jmetalpy.

```python
from moo.lz import LZ1, LZ2, LZ3, LZ4, LZ5, LZ6, LZ7, LZ8, LZ9
from moo.lz_jmetalpy import LZ09_F1, LZ09_F2, LZ09_F3, LZ09_F4, LZ09_F5, LZ09_F6, LZ09_F7, LZ09_F8, LZ09_F9
from helper import plot_front

from plotly.subplots import make_subplots
import plotly.graph_objects as go

n_var = 10

def create_subplot(plot_of_my_implementation, plot_of_jmetalpy, title) -> go.Figure:
    fig = make_subplots(rows=1, cols=2, subplot_titles=("own", "jmetalpy"))
    for trace in plot_of_my_implementation.data:
        fig.add_trace(trace, row=1, col=1)
    for trace in plot_of_jmetalpy.data:
        fig.add_trace(trace, row=1, col=2)
    fig.update_layout(title_text=title)
    return fig

fig = create_subplot(
    plot_front(problem=LZ1(n_var=n_var)),
    plot_front(problem=LZ09_F1(n_var=n_var)),
    "LZ1"
)
fig.show()

fig = create_subplot(
    plot_front(problem=LZ2(n_var=n_var)),
    plot_front(problem=LZ09_F2(n_var=n_var)),
    "LZ2"
)
fig.show()

fig = create_subplot(
    plot_front(problem=LZ3(n_var=n_var)),
    plot_front(problem=LZ09_F3(n_var=n_var)),
    "LZ3"
)
fig.show()

fig = create_subplot(
    plot_front(problem=LZ4(n_var=n_var)),
    plot_front(problem=LZ09_F4(n_var=n_var)),
    "LZ4"
)
fig.show()

fig = create_subplot(
    plot_front(problem=LZ5(n_var=n_var)),
    plot_front(problem=LZ09_F5(n_var=n_var)),
    "LZ5"
)
fig.show()

plot_front(problem=LZ6(n_var=n_var)).show()
plot_front(problem=LZ09_F6(n_var=n_var)).show()

fig = create_subplot(
    plot_front(problem=LZ7(n_var=n_var)),
    plot_front(problem=LZ09_F7(n_var=n_var)),
    "LZ7"
)
fig.show()

fig = create_subplot(
    plot_front(problem=LZ8(n_var=n_var)),
    plot_front(problem=LZ09_F8(n_var=n_var)),
    "LZ8"
)
fig.show()

fig = create_subplot(
    plot_front(problem=LZ9(n_var=n_var)),
    plot_front(problem=LZ09_F9(n_var=n_var)),
    "LZ9"
)
fig.show()
```
