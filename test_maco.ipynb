# p-norm variation

```python
import numpy as np
from moo.maco import MACO
from helper import run_test, plot_front

problem=MACO(n_var=10, p=-np.inf)
res = run_test(problem)
plot_front(problem, res.F, "MACO p=-inf").show()

problem=MACO(n_var=10, p=-10)
res = run_test(problem)
plot_front(problem, res.F, "MACO p=-10").show()

problem=MACO(n_var=10, p=-2)
res = run_test(problem)
plot_front(problem, res.F, "MACO p=-2").show()
```
